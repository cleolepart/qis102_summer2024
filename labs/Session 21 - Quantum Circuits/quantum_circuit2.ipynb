```python
# quantum_circuit2.ipynb

from IPython.display import display
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_distribution
from qiskit_aer import AerSimulator

# Cell 1 - Circuit to produce only two states (63 or 64 decimal)
qc = QuantumCircuit(7)

# Implement the circuit
qc.x(0)  # Set the first 6 qubits to |1>
qc.x(1)
qc.x(2)
qc.x(3)
qc.x(4)
qc.x(5)
qc.h(6)  # Apply Hadamard gate to the 7th qubit
qc.cx(6, 0)  # CNOT with control on qubit 6 and target on qubit 0
qc.cx(6, 1)  # CNOT with control on qubit 6 and target on qubit 1
qc.cx(6, 2)  # CNOT with control on qubit 6 and target on qubit 2
qc.cx(6, 3)  # CNOT with control on qubit 6 and target on qubit 3
qc.cx(6, 4)  # CNOT with control on qubit 6 and target on qubit 4
qc.cx(6, 5)  # CNOT with control on qubit 6 and target on qubit 5

# Measurement (optional, but useful for visualization)
qc.measure_all()

# Do not modify the following code
display(qc.draw(output="mpl"))

backend = AerSimulator()
qc_transpiled = transpile(qc, backend)
result = backend.run(qc_transpiled, shots=100_000).result()
counts = result.get_counts(qc)
display(plot_distribution(counts))
```
